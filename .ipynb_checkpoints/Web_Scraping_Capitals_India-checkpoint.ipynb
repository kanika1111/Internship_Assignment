```python
#import the library used to query a website
import urllib3
import warnings
warnings.filterwarnings('ignore')
```

```python
#specify the url
wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
```

```python
#Query the website and return the html to the variable 'page'
http = urllib3.PoolManager()
response = http.request('GET', wiki)
page= response.data
```

```python
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page)
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
all_links=soup.find_all('a')
for link in all_links:
    print (link.get("href"))
```

```python
right_table = soup.find('table',class_='wikitable sortable plainrowheaders')
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]

for row in right_table.findAll("tr"):    
    states=row.findAll('th') #To store second column data 
    if len(states)==2: #Only extract table body not heading
        cells=row.findAll('td') #To store second column data
        A.append(states[0].find(text=True))
        B.append(states[1].find(text=True))
        C.append(cells[0].find(text=True))
        D.append(cells[1].find(text=True))
        E.append(cells[2].find(text=True))
        F.append(cells[3].find(text=True))
        G.append(cells[4].find(text=True))
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame()
df['Number']=A
df['State/UT']=B
df['Admin_Capital']=C
df['Legislative_Capital']=D
df['Judiciary_Capital']=E
df['Year_Capital']=F
df['Former_Capital']=G
```

```python
df.head()
```

```python
df.to_csv("Indian_territory_info.csv", encoding = 'utf-8')
```
